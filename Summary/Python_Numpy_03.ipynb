#### 슬라이싱

 - 리스트, 문자열 slicing과 동일한 개념으로 사용
 - ,를 사용하여 각 차원 별로 슬라이싱 가능
 
#### 1차원 벡터 슬라이싱

```python
x = np.arange(10)
```

```python
x[1:7]
```

#### 2차원 행렬 슬라이싱

```python
x = np.arange(10).reshape(2, 5)
x
```

```python
x[1, 1:3]
```

#### 3차원 텐서 슬라이싱

```python
x = np.arange(36).reshape(3, 4, 3)
x
```

```python
x[:2, :3, :2]
```

#### ravel, np.ravel

 - 다차원배열을 1차원으로 변경
 - 'order'파라미터
     - 'C' - row 우선변경
     - 'F' - column 우선변경

```python
x = np.arange(15).reshape(3, 5)
x
```

```python
x.ravel()
```

```python
np.ravel(x)
```

```python
temp = x.ravel()
temp
```

```python
temp[0] = 100
print(temp)
print(x)
```

#### flatten

 - 다차원 배열을 1차원으로 변경
 - ravel 과의 차이점 : copy를 생성하여 변경함(즉, 원본데이터가 아닌 복사본을 반환)
 - 'order' 파라미터
     - 'C' - row 우선변경
     - 'F' - column 우선 변경

```python
y = np.arange(15).reshape(3, 5)
y
```

```python
y.flatten()
```

```python
y.flatten(order='F')
```

```python
temp = y.flatten()
```

```python
temp[0] = 100
print(temp)
print(y)
```

#### reshape 함수

 - array의 shape을 다른 차원으로 변경
 - 주의할 점은 reshape한 후의 결과의 전체 원소 개수와 이전 개수가 같아야 가능
 - 사용 
     * 예) 이미지 데이터 벡터화 - 이미지는 기본적으로 2차원 혹은 3차원(RGB)이나 트레이닝을 위해 1차원으로 변경하여 사용 됨

```python
x = np.arange(36)
x
```

```python
x.shape
```

```python
x.ndim
```

```python
x.reshape(6, 6)
```

```python
x.reshape(6, -1)
```

```python
x = np.arange(15).reshape(3, 5)
y = np.random.rand(15).reshape(3, 5)
print(x)
print(y)
```
