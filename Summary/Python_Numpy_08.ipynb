#### np.linalg.inv

 - 역행렬을 구할 때 사용
 - 모든 차원의 값이 같아야 함

```python
x = np.random.rand(3, 3)  # 정방행렬만 가능
x
```

```python
x @ np.linalg.inv(x)
```

```python
np.matmul(x, np.linalg.inv(x))
```

#### np.linalg.solve

 - $Ax = B$ 형태의 선형대수식 솔루션을 제공
 - ex) 호랑이와 홍합의 합 : 25 호랑이 다리와 홍합 다리의 합은 64
     - $x+y = 25$
     - $2x + 4y = 64$

```python
A = np.array([[1, 1], [2, 4]])
A
```

```python
B = np.array([25, 64])
B
```

```python
x = np.linalg.solve(A, B)
x
```

```python
np.allclose(A@x, B)  # 실수 값 비교
```
