#### any, all 함수

 - any : 특정 조건을 만족하는 것이 하나라도 있으면 True, 아니면 False
 - all : 모든 원소가 특정 조건을 만족한다면 True, 아니면 False

```python
z = np.random.randn(10)
z
```

```python
np.any(z > 0)
```

```python
np.all(z != 0)
```

```python
z > 0
```

#### where 함수

 - 조건에 따라 선별적으로 값을 선택 가능
 - 사용 예) 음수인 경우는 0, 나머지는 그대로 값을 쓰는 경우

```python
z = np.random.randn(10)
z
```

```python
np.where(z > 0, z, 0)  # 조건, True, False
```
