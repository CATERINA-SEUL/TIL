### random sub module

#### rand 함수

 - 0, 1 사이의 분포로 랜덤한 ndarray 생성


```python
np.random.rand(2, 2)
```

#### randn 함수 
 - n: normal distribution(정규분포)
 - 정규분포로 샘플링 된 랜덤 ndarray 생성

```python
np.random.randn(2, 2)
```

#### randint 함수

 - 특정 정수 사이에서 랜덤하게 샘플링

```python
np.random.randint(2, 100, (6, 3))
```

#### seed 함수

 - 랜덤한 값을 동일하게 다시 생성하고자 할 때 사용

```python
np.random.seed(100)
```

```python
np.random.randn(3, 4)
```

#### choice
 - 주어진 1차원 ndarray로부터 랜덤으로 샘플링
 - 정수가 주어진 경우 np.arange(해당숫자)로 간주

```python
np.random.choice(100, (3, 4))
```

```python
x = np.array([1, 2, 3, 1.5, 2.6, 4.9])
np.random.choice(x, (2, 2))
```

#### 확률분포에 따른 ndarray 생성

 - uniform
 - normal 등

```python
np.random.uniform(1.0, 3.0, (4, 5))
```

```python
np.random.normal(1.0, 3.0, (4, 5))
```

### indexing 

 - python list 와 동일한 개념으로 사용
 - , 를 사용하여 각 차원의 인덱스에 접근가능
 
 #### 1차원 벡터 인덱싱

```python
x = np.arange(1, 10)
x
```

```python
x[0]
```

```python
x[::-1]
```

#### 2차원 행렬 인덱싱

```python
x = np.arange(10).reshape(2, 5)
print(x)
```

```python
x[0, 3]
```

#### 3차원 텐서 인덱싱

```python
x = np.arange(36).reshape(3, 4, 3)
x
```

```python
x[1, 3, 2]
```
