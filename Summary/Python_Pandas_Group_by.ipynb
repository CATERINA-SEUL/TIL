#### groupby

 - 아래의 3 단계를 적용하여 데이터를 그룹화(groupping) 
      - SQL의 group by와 개념적으로는 동일, 사용법은 유사
     
    - 데이터 분할
    - operation 적용
    - 데이터 병합

#### GroupBy groups 속성

 - 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
df = pd.read_csv('./train.csv')
df.head()
```

```python
class_group = df.groupby('Pclass')
class_group
```

```python
class_group.groups
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

#### groupping 함수

 - 그룹 데이터에 적용 가능한 통계 함수 (NaN은 제외하여 연산)
 - count : 데이터 개수
 - sum : 데이터의 합
 - mean, std, var : 평균, 표준편차, 분산 
 - min, max : 최소, 최대값

```python
class_group.count()
```

```python
gender_group.count()
```

```python
class_group.mean()['Survived']
```

```python
class_group.min()
```

```python
df.groupby(['Pclass', 'Sex']).mean().loc[(2, 'female')]
```

```python
df.groupby(['Pclass', 'Sex']).mean()['Survived']
```

#### index 를 이용한 group by

 - index 가 있는 경우, groupby 함수에 level 사용 가능
     - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
     
 - set_index 함수
     - column 데이터를 index 레벨로 변경
     
 - reset_index 함수
     - 인덱스 초기화

```python
df.head()
```

```python
df.set_index('Pclass').head()
```

```python
df.set_index(['Pclass', 'Age']).head()
```

```python
df.set_index('Age').groupby(level=0).mean().tail()
```
