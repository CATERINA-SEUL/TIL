#### Data Frame

 - Series가 1차원이라면 DataFrame은 2차원으로 확대된 버전
 - excel spreadsheet 라고 생각하면 이해하기 쉬움
 - 2차원이기 때문에 인덱스가 row, column으로 구성됨
     - row는 각 개별 데이터를, column 은 개별 속성을 의미
 - Data Analysis, Machine Learning 에서 data 변형을 위해 가장 많이 사용

```python
train_data = pd.read_csv("./train.csv")
train_data.tail()
```

#### Data Frame 생성하기

 - 일반적으로 분석을 위한 데이터는 다른 데이터 소스(database, 외부파일)을 통해 dataframe을 생성
 - 여기서는 실습을 통해 dummy 데이터를 생성하는 방법을 다룰 예정

#### dictionary로 부터 생성하기

 - dict의 key $\rightarrow$ column

```python
data = {'a':[1,2,3],'b':[4,5,6],'c':[6,7,8]} #column
pd.DataFrame(data, index=['x','y','d']) #index
```

```python
data = {'a':100,'b':200,'c':300} #column
pd.DataFrame(data, index=[0,1,2]) #index
```

#### Series로 부터 생성하기  

 - 각 Series의 인덱스 $\rightarrow$ column

```python
a = pd.Series([100,200,300],['a','b','c'])
b = pd.Series([101,201,301],['a','b','c'])
c = pd.Series([103,203,303],['a','b','c'])

pd.DataFrame([a,b,c])
```
