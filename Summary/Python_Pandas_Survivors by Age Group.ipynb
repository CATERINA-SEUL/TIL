```python
import math
```

```python
df = pd.read_csv('/Users/caterina/Dropbox/Study/Fast Campus/Data Science School/DSS1/02_numpy_pandas/02_numpy_pandas/datas/train.csv')
df.head()
```

```python
def Age_Categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
df.set_index('Age').groupby(Age_Categorize).mean()['Survived']
```

#### MultiIndex를 이용한 groupping

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).mean()['Age'].head()
```

#### aggregate(집계) 함수 사용하기

 - groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능

```python
df.set_index(['Pclass', 'Sex']).groupby(
    level=[0, 1]).aggregate([np.mean, np.sum, np.max])
```

#### transform 함수

 - groupby 후 transform 함수를 사용하면 원래의 index를 유지한 상태로 통계함수를 적용
 - 전체 데이터의 집계가 아닌 각 그룹에서의 집계를 계산
 - 따라서 새로 생성된 데이터를 원본 dataframe과 합치기 쉬움

```python
df.groupby('Pclass').mean()
```

```python
df.groupby('Pclass').transform(np.mean).head()
```

```python
df['Age2'] = df.groupby('Pclass').transform(np.mean)['Age']
df.tail()
```

```python
df.groupby(['Pclass', 'Age']).mean().tail()
```

```python
df['Age3'] = df.groupby(['Pclass', 'Sex']).transform(np.mean)['Age']
df.tail()
```
