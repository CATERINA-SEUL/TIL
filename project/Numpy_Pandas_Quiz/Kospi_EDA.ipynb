#### (3) 2019년 12월 08일 이후의 날짜별 평균 상승률을 소수 셋째자리까지 리스트의 딕셔너리 형태로 데이터를 출력하세요.
 -  (dt : 날짜, cr : 상승률)
 - (출력 포멧 예시 : [{'dt': 20191109, 'cr': -1.234}, {'dt': 20191110, 'cr': 5.512}])


```python
dt1 = df3[df3["dt"] > 20191208]
dt1 = dt1.sort_values(["dt"])
dt1.head(10)
```

```python
df4 = dt1.pivot_table("cr", "dt", aggfunc='mean').reset_index()
df4 = round((df4), 3)
df4
```

```python
df5 = df4.to_dict("recode")
df5
```

#### (5) 2019년 11월 30일에서 2019년 12월 5일까지 일별 상승률 상위 5개 회사의 평균 상승율을 소수점 셋째 자리까지 출력하세요. (cr : 상승률)

```python
dt3 = df6[(df6["dt"] >= 20191130) & (df6["dt"] <= 20191205)]
df7 = dt3.groupby(["dt", "name", 'cr'], sort=True).size().reset_index()
df7 = df7.sort_values("cr", ascending=False)
df7 = round((df7), 3)
df7.reset_index(drop=True, inplace=True)
df7
```

```python
df8 = df7.loc[:5]
df8 = df8.drop(df8.index[4])
df9 = df8["cr"].apply("mean")
df10 = round((df9), 3)
display(df8)
print("상위 5개 회사 평균 상승율 : ", df10)
```

#### (7) NAVER와 한국전력의 종가 데이터의 상관계수를 소수둘째자리까지 출력하세요. (ncv: 종가)
